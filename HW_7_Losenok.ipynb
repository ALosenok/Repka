```python
import math
# Task 1. Money Money Money
class Money:
    def __init__(self, rub = 0, cent = 0):
        self.__rub = rub
        self.__cent = cent

    # A getter to return rubles value
    @property
    def rub(self):
        return self.__rub

    # A setter to define new cents value
    @rub.setter
    def rub(self, rub):
        if rub >= 0:
          self.__rub = rub
        else:
          print("The value can´t be negative")

    # A getter to return cents value
    @property
    def cent(self):
        return self.__cent

    # A setter to define new cents value
    @cent.setter
    def cent(self, cent):
        if cent >= 100:
          self.__cent = cent % 100
          self.__rub += cent // 100
        elif 99 > cent >= 0:
          self.__cent = cent
        else:
          print("The value can´t be negative")

    # A method that returns actual value
    def value(self):
            return print(str(self.__rub) + "," + str(self.__cent))

    # A method that adds more money
    def load(self, rub = 0, cent = 0):
        if rub < 0:
            print("Rub value can´t be negative")
        else:
            self.__rub += rub + (self.__cent + cent) // 100
        if cent < 0:
            print("Cent value can´t be negative")
        else:
            self.__cent = (self.__cent + cent) % 100

    # A method that removes money
    def unload(self, rub = 0, cent = 0):
        if rub < 0:
            print("Rub value can´t be negative")
        else:
            self.__rub -= rub - (self.__cent + cent) // 100
        if cent < 0:
            print("Cent value can´t be negative")
        else:
            self.__cent = (self.__cent + cent) % 100

    # A method that alterates value using
    def alt(self, rub = 0, cent = 0):
            self.__rub += rub + (self.__cent + cent) // 100
            self.__cent = (self.__cent + cent) % 100

    # A method that divides the value by integer divisor
    def div_int(self, divisor = 1):
        if divisor == 0:
            print("Divisor can´t be equal to 0")
        elif divisor == 1:
            print("Set a divisor please")
        else:
            n_rub = math.ceil((self.__rub * 100 + self.__cent) / divisor) // 100
            n_cent = math.ceil((self.__rub * 100 + self.__cent) / divisor) % 100
            self.__rub = n_rub
            self.__cent = n_cent

    # A method that divides the value by float divisor
    def mult(self, mult = 1):
        if mult < 0:
            print("Multiplier can´t be negative")
        elif mult == 1:
            print("Set a multiplier please")
        else:
            temp_rub = math.ceil((self.__rub * 100 + self.__cent) * mult) // 100
            temp_cent = math.ceil((self.__rub * 100 + self.__cent) * mult) % 100
            self.__rub = temp_rub
            self.__cent = temp_cent

    # A method that compares values of different objects
    def compare(self, other):
        dif = round(
        (self.__rub * 100 + self.__cent) / (other.__rub * 100 + other.__cent), 2)
        if dif < 1:
            print(f"The first balance is less than the second in {round(1 / dif, 2)} times")
        elif dif > 1:
            print(f"The first balance is greater than the second in {dif} times")
        else:
            print(f"The first balance equal to the second")






sotka = Money(5, 33)
#sotka.rub = 50
sotka.rub
sotka.value()
sotka.load(0,77)
sotka.value()
sotka.alt(-1,-110)
sotka.value()
sotka.div_int(2.4)
sotka.value()
sotka.mult(2.1)
sotka.value()
dvusotka = Money(3,0)
dvusotka.value()
sotka.compare(dvusotka)
#почему rub работает без (), a value требует ()?
#Нужно ли прописывать описание в методах как в прочих кастомных функциях
```
