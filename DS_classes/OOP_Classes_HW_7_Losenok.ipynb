```python
import math
# Task 1. Money Money Money
class Money:
    def __init__(self, rub: int = 0, cent: int = 0):
        """
        A method that creates an object within class Money

        :param rub: an integer that determines a value of rubles
        :param cent: an integer that determines a values of cents
        :return: an object class Money
        """
        self.__rub: int = rub
        self.__cent: int = cent

    # A getter to return rubles value
    @property
    def rub(self) -> int:
      """
      A getter method that return a value of rubles in an object

      :return: an integer with a value of rubles
      """
      return self.__rub

    # A setter to define new cents value
    @rub.setter
    def rub(self, rub: int):

      """
      A setter methods that sets a new value of rubles in an object

      :param rub: an integer that determines a value of rubles
      """

      if rub >= 0:
          self.__rub = rub
      else:
          print("The value can´t be negative")

    # A getter to return cents value
    @property
    def cent(self) -> int:
        """
        A getter method that return a value of cents in an object

        :return: an integer with a value of cents
        """
        return self.__cent

    # A setter to define new cents value
    @cent.setter
    def cent(self, cent: int):
        """
        A setter methods that sets a new value of rubles in an object

        :param rub: an integer that determines a value of rubles
        """

        if cent >= 100:
          self.__cent = cent % 100
          self.__rub += cent // 100
        elif 99 > cent >= 0:
          self.__cent = cent
        else:
          print("The value can´t be negative")

    # A method that returns actual value
    def value(self) -> str:
        """
        A method that returns values of rubles and cents separated by coma

        :return: a string representing a value of rubles and cents
        """
        return str(self.__rub) + "," + str(self.__cent)

    # A method that adds more money
    def load(self, rub: int = 0, cent: int = 0):
        """
        A method that adds values to the attributes of the object

        :param rub: a positive integer defining a value of rubles to be added
        :param cent: a positive integer defining a value of cents to be added
        """

        if rub < 0:
            print("Rub value can´t be negative")
        else:
            self.__rub += rub + (self.__cent + cent) // 100
        if cent < 0:
            print("Cent value can´t be negative")
        else:
            self.__cent = (self.__cent + cent) % 100

    # A method that removes money
    def unload(self, rub = 0, cent = 0):
        """
        A method that subtracts values from the attributes of the object

        :param rub: a positive integer defining a value of rubles to be subtracted
        :param cent: a positive integer defining a value of cents to be subtracted
        """

        if rub < 0:
            print("Rub value can´t be negative")
        else:
            self.__rub -= rub - (self.__cent + cent) // 100
        if cent < 0:
            print("Cent value can´t be negative")
        else:
            self.__cent = (self.__cent + cent) % 100

    # A method that alterates value using
    def alt(self, rub = 0, cent = 0):
            """
            A method that alterate values from the attributes of the object
            depending on sign of the integers

            :param rub: a positive integer defining a value of rubles to be
            added or subtracted
            :param cent: a positive integer defining a value of cents to be
            added or subtracted
            """

            self.__rub += rub + (self.__cent + cent) // 100
            self.__cent = (self.__cent + cent) % 100

    # A method that divides the value by integer divisor
    def div_int(self, divisor: int or float = 1):
        """
        A method that define new values of attributes in result of division of
        the object

        :param divisor: an integer or float value that the object is divided by
        """

        if divisor == 0:
            print("Divisor can´t be equal to 0")
        elif divisor == 1:
            print("Set a divisor please")
        else:
            n_rub = math.ceil((self.__rub * 100 + self.__cent) / divisor) // 100
            n_cent = math.ceil((self.__rub * 100 + self.__cent) / divisor) % 100
            self.__rub = n_rub
            self.__cent = n_cent


    def mult(self, mult: int = 1):
        """
        A method that define new values of attributes in result of multiplication
        of the object

        :param mult: an integer or float value that the object is multiplied by
        """

        if mult < 0:
            print("Multiplier can´t be negative")
        elif mult == 1:
            print("Set a multiplier please")
        else:
            temp_rub = math.ceil((self.__rub * 100 + self.__cent) * mult) // 100
            temp_cent = math.ceil((self.__rub * 100 + self.__cent) * mult) % 100
            self.__rub = temp_rub
            self.__cent = temp_cent

    # A method that compares values of different objects
    def compare(self, other) -> str:
        """
        A method that compare two objects

        :returns: a string with result of camparison and a value of times one object
        differ of another
        """

        dif = round(
        (self.__rub * 100 + self.__cent) / (other.__rub * 100 + other.__cent), 2)
        if dif < 1:
            print(f"The first balance is less than the second in {round(1 / dif, 2)} times")
        elif dif > 1:
            print(f"The first balance is greater than the second in {dif} times")
        else:
            print(f"The first balance equal to the second")


    def __eq__(self, other) -> bool:
        if not isinstance(other, Money):
            return f"{other} is not Money object"
        else:
            self.__rub * 100 + self.__cents == other.__rub * 100 + other.__cents

    def __lt__(self, other) -> bool:
        if not isinstance(other, Money):
            return f"{other} is not Money object"
        else:
            self.__rub * 100 + self.__cents < other.__rub * 100 + other.__cents

    def __le__(self, other) -> bool:
        if not isinstance(other, Money):
            return f"{other} is not Money object"
        else:
            self.__rub * 100 + self.__cents <= other.__rub * 100 + other.__cents

    def __gt__(self, other) -> bool:
        if not isinstance(other, Money):
            return f"{other} is not Money object"
        else:
            self.__rub * 100 + self.__cents > other.__rub * 100 + other.__cents

    def __ge__(self, other) -> bool:
        if not isinstance(other, Money):
            return f"{other} is not Money object"
        else:
            self.__rub * 100 + self.__cents >= other.__rub * 100 + other.__cents



sotka = Money(5, 33)
#sotka.rub = 50
sotka.rub
sotka.value()
sotka.load(0,77)
sotka.value()
sotka.alt(-1,-110)
sotka.value()
sotka.div_int(2.4)
sotka.value()
sotka.mult(2.1)
sotka.value()
dvusotka = Money(3,0)
dvusotka.value()
sotka.compare(dvusotka)

```

```python
class Matrix_1:
    def __init__(self):
        self.__mtx: list = []
        self.__rows: int = 0
        self.__lines: int = 0

    @property
    def mtx(self):
        """
        A getter method that returns each element of the matrix printed in a
        separated line
        """
        for i in self.__mtx:
            print(i)


    @property
    def lines(self):
        return self.__lines


    @property
    def rows(self):
        return self.__rows


    def gen(self, *lists: list):
        """
        A method that generates a final matrix from a number of list arguments

        :param *lists: a number of lists that will be included in the matrix
        :return: a generated Matrix_1 object with n(number of lists) rows and
        m(number of elements in the largest list) columns
        """
        max_len = 0

        for lst in lists:         # Define the max length of the lists
            if len(lst) > max_len:
                max_len = len(lst)

        all_lst = 1 # Flag to check if all args are lists
        all_int = 1 # Flag to check if all elements of lists are integers
        num_rows = 0 # Define a final number of rows

        for lst in lists:
            num_rows += 1
            if isinstance(lst, list): #C hecking if all the arguments are lists
                for i in lst:
                    if isinstance(i, int): # Checking if all elements in list are ints
                        continue
                    else:
                        print(f"Element {i} in {lst} is not integer")
                        all_int = 0
                        break
            else:
                print("All arguments must be lists")
                all_lst = 0
                break

        if all_lst == 1 and all_int == 1:
            for lst in lists:
                for dif in range(max_len - len(lst)): #Adding zero if the list has different length
                    lst.append(0)
                self.__mtx.append(lst)
            self.__lines = max_len # Define number of lines
            self.__rows = num_rows # Define number of rows


    def __getitem__(self, key: tuple) -> int: # Creating access to elements of an object
        row, line = key
        return self.__mtx[row][line]


    def __setitem__(self, key: tuple, val: int):
        row, line = key
        self.__mtx[row][line] = val


    def multiply(self, val: int or float):
        """
        A method for multiply an object by a value

        :param val: an integer or float that the object will be multiplied by
        """

        for i in range(self.__rows):
            for j in range(self.__lines):
                self.__mtx[i][j] *= val


    def add_mtx(self, other):
        """
        A method that sums up two matricies
        """

        if self.__rows == other.__rows and self.__lines == other.__lines:
            for i in range(self.__rows):
                for j in range(self.__lines):
                    self.__mtx[i][j] += other.__mtx[i][j]
        else:
            print("The matricies which differ in dimensions can´t be added!")

    def mult_vect(self, other):
        """
        A method that multiplies two matrices
        """

        if self.__lines == other.__rows and other.__lines == 1:
            temp_lst = []
            tmp_sum = 0
            temp_lst_2 = []
            for i in range(self.__rows + 1):
                temp_lst_2.append(tmp_sum)
                temp_lst.append(temp_lst_2)
                temp_lst_2 = []
                tmp_sum = 0
                for j in range(other.__rows):
                  if i == self.__rows:
                      break
                  else:
                      tmp_sum += self[i, j] * other[j, 0]

            temp_lst = temp_lst[1:]

        elif self.__rows == other.__lines and other.__rows == 1 and self.__lines == 1:
              temp_lst = []
              temp_lst_2 = []
              for i in range(self.__rows + 1):
                  temp_lst.append(temp_lst_2)
                  temp_lst_2 = []
                  for j in range(other.__lines):
                    if i == self.__rows:
                        break
                    else:
                        temp_sum = self[i, 0] * other[0, j]
                        temp_lst_2.append(temp_sum)
              temp_lst = temp_lst[1:]

        else:
                raise ValueError ("Matrices with this dimensions can´t be multiplied")

        return temp_lst


    def mult_mtx(self, other):
        """
        A function that multiplies two matricies
        :result: a nested list with multiplied values
        """

        if self.__lines == other.__rows:
            lst = []
            for i in range(self.__rows):
                #print(i)
                lst_2 = []
                for k in range(other.__lines):
                    #print(k)
                    tmp_sum = 0
                    for j in range(other.__rows):
                        #print(j)
                        tmp_sum += self[i, j] * other[j, k]
                    lst_2.append(tmp_sum)
                lst.append(lst_2)
        else:
            raise ValueError("The matrices with these dimensions can´t be multiplied")

        return lst



m2 = Matrix_1()
m2.gen([1,0], [1,1, 1, 1, 1], [4, 8, 25])
m3 = Matrix_1()
m3.gen([1,0], [1,1, 1, 1, 1], [4, 8, 25])
m2.mtx
m2.lines
print(m2[2,1])
m2.multiply(2)
m2.mtx
m2[0, 2] = 1
m2.mtx
m2.add_mtx(m3)
m2.mtx
```
